## O que é Mineração de Dados?

Mineração de dados é área de estudos do métodos para descoberta de padrões em bases de dados, de modo que, a partir dos padrões descobertos, se estruture um conhecimento útil para um processo de tomada de decisão.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/2/25/Cart_tree_kyphosis.png/800px-Cart_tree_kyphosis.png)

A mineração de dados consiste assim na aplicação de técnicas, implementadas por meio de algoritmos computacionais, capazes de receber como entrada registros de fatos ocorridos no mundo real e devolver como saída a descrição de um padrão de comportamento, que pode ser expressa, por exemplo, como uma regra de associação, uma função de mapeamento ou uma modelagem.

As soluções e oportunidades proporcionadas pela mineração de dados variam conforme o tipo de dado disponível e a natureza do conhecimento necessário. Assim, a área de mineração de dados é comumente dividida em tarefas, as quais ajudam a entender como situar um problema real junto aos diferentes algoritmos disponíveis e também que tipo de conhecimento é possível descobrir.

## Tarefas de Mineração de Dados

Tarefas de mineração de dados são divididas em **preditivas** e **descritivas**.

Tarefas **preditivas** usam os valores dos atributos descritivos para predizer valores futuros ou desconhecidos de outros atributos de interesse.

Tarefas **descritivas** têm o objetivo de encontrar padrões que descrevem os dados de maneira que o ser humano possa interpretar.

No conjunto de tarefas preditivas, estão inseridos os problemas de **classificação** e **regressão**.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/7/72/SVM_margin.png/300px-SVM_margin.png)

No conjunto de tarefas descritivas, estão inseridos os problemas de **agrupamento**, **sumarização**, **modelagem de dependências** e **detecção de desvios**.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/c/c8/Cluster-2.svg/220px-Cluster-2.svg.png)
